```python
df1 = sht[['seas_id', 'season', 'player_id', 'player', 'pos', 'age',
       'experience', 'tm', 'g', 'mp', 'avg_dist_fga', 'percent_fga_from_x10_16_range',
       'percent_fga_from_x16_3p_range', 'fg_percent_from_x10_16_range',
       'fg_percent_from_x16_3p_range']]
df2 = tot[['seas_id', 'fg', 'fga', 'fg_percent', 'x2p', 'x2pa', 'x2p_percent', 'pts']]

merged = pd.merge(df1, df2, on='seas_id', how='inner')
merged.head()
```

```python
merged['midrange fga'] = merged['fga'] * (merged['percent_fga_from_x10_16_range'] + merged['percent_fga_from_x16_3p_range'])
merged['midrange fg%'] = merged['fg_percent_from_x10_16_range'] * merged['fg_percent_from_x16_3p_range']
merged['midrange pts'] = merged['midrange fga'] * merged['midrange fg%']

merged.head()
```

```python
merged['mid score'] = (merged['avg_dist_fga'] * 1.75) + (merged['midrange pts'] * 0.5) * ((merged['midrange fga'] / merged['x2pa']) * 1.75) * (merged['midrange fg%'] * 3)

sorted = merged.sort_values(by='mid score', ascending=False)

sorted.head(10)
```

```python
# Dropping duplicates to keep individul player best season
best_player_szn = sorted.drop_duplicates(subset='player', keep='first')

col = ['player', 'tm', 'season', 'mid score','midrange pts', 'midrange fga', 'midrange fg%', 'avg_dist_fga']
best_player_szn[col].head(15)
```
