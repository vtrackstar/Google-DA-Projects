```python
glr = pd.read_csv('/kaggle/input/nflstatistics/Game_Logs_Wide_Receiver_and_Tight_End.csv')
crr = pd.read_csv('/kaggle/input/nflstatistics/Career_Stats_Receiving.csv')

print(glr.columns)
print(crr.columns)
```

```python
pd.set_option('display.max_columns', None)
cols = ['Games Played', 'Receptions', 'Receiving Yards', 'Yards Per Reception',
        'Yards Per Game', 'Longest Reception', 'Receiving TDs',
        'Receptions Longer than 20 Yards', 'Receptions Longer than 40 Yards',
        'First Down Receptions', 'Fumbles']

crr[cols] = crr[cols].apply(pd.to_numeric, errors='coerce')
crr1 = crr.fillna(0)
crr1.head()
```

```python
crr1['score'] = (((crr1['Receptions']) / (crr1['Games Played'])) * 0.5) + (crr1['Yards Per Game'] * 0.25) + (crr1['Yards Per Reception'] * 0.5) + (crr1['Receiving TDs'] * 6) + (crr1['Fumbles'] * -4)
sortcrr = crr1.sort_values(by='score', ascending=False)
sortcrr.head()
```

```python
bestcar = sortcrr.drop_duplicates(subset='Name', keep='first')
bestcar.head(10)
```

```python
df1 = bestcar[~bestcar['Position'].isin(['TE', 'RB'])]
df2 = bestcar[bestcar['Position'] == 'TE']
df3 = bestcar[bestcar['Position'] == 'RB']
```
