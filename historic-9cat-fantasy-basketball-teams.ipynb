```python
cat9 = tm[['season', 'g', 'team', 'abbreviation', 'pts', 'trb', 'ast', 'x3p', 'blk', 'stl', 'fg_percent', 'ft_percent', 'tov']]

c9_clean = cat9.dropna()
c9_cleaned = c9_clean.copy()
c9_cleaned.head()
```

```python
c9_cleaned['cat9_finish'] = c9_cleaned['pts'] + c9_cleaned['trb'] + c9_cleaned['ast'] + c9_cleaned['x3p'] + c9_cleaned['blk'] + c9_cleaned['stl'] + c9_cleaned['fg_percent'] + c9_cleaned['ft_percent'] - c9_cleaned['tov']
c9_cleaned['per_game_avg'] = c9_cleaned['cat9_finish'] / c9_cleaned['g']

sort = c9_cleaned.sort_values(by=['per_game_avg'], ascending=False)

# Display the result
sort.head()
```

```python
best_szn = sort.drop_duplicates(subset='team', keep='first')
best_szn.head(10)
```

```python
leaders = sort.groupby('season').first().reset_index()
leaders.head(77)
```

```python
per_game = leaders.copy()
per_game['pts'] = per_game['pts'] / per_game['g']
per_game['trb'] = per_game['trb'] / per_game['g']
per_game['ast'] = per_game['ast'] / per_game['g']
per_game['x3p'] = per_game['x3p'] / per_game['g']
per_game['blk'] = per_game['blk'] / per_game['g']
per_game['stl'] = per_game['stl'] / per_game['g']
per_game['tov'] = per_game['tov'] / per_game['g']
per_game.head(50)
```
