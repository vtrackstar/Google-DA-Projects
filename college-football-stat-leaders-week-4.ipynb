```python
print(qb.head())
print(qb.columns)
print(rb.head())
print(rb.columns)
print(wr.head())
print(wr.columns)
```

```python
qb['Score'] = (qb['Cmp%'] * 2.5) + (qb['Y/G'] * 1.5) + (qb['TD'] * 3) + (qb['Int'] * -3) + (qb['Y/A'] * 2)
rb['Score'] = (rb['RAvg'] * 1.5) + (rb['RTD'] * 3) + ((rb['RYds'] / rb['G']) * 1.5)
wr['Score'] = (wr['Rec'] * 1.5) + (wr['RecAvg'] * 2) + (wr['RecTD'] * 3) + ((wr['RecYds'] / wr['G']) * 2)

rb['Y/G'] = rb['RYds'] / rb['G']
wr['Y/G'] = wr['RecYds'] / wr['G']

qb_sort = qb.sort_values(by='Score', ascending=False)
rb_sort = rb.sort_values(by='Score', ascending=False)
wr_sort = wr.sort_values(by='Score', ascending=False)

print(qb_sort.head())
print(rb_sort.head())
print(wr_sort.head())
```
