```python
qb = pd.read_csv('/kaggle/input/nfl-qb-stats-1970-2022/NFL QB Stats.csv')
qb.columns
```

```python
qb.head()
```

```python
calc = qb.copy()
calc['20-40'] = calc['20+'] - calc['40+']
calc['%Deep'] = calc['20+'] / calc['Cmp']
calc.head()
```

```python
calc10 = calc[calc['Att'] >= 500]
calc1 = calc10.copy()
calc1['Score'] = (calc1['%Deep'] * 225) + (calc1['20+'] * 0.9) + (calc1['Yds/Att'] * 0.3) + (calc1['Att'] * 0.1) + (calc1['INT'] * -0.15) + (calc1['TD'] * 0.25) + (calc1['Sck'] * -0.25)
```

```python
sortdeep = calc1.sort_values(by='Score', ascending=False)
sortdeep.head(10)
```

```python
bestszn = sortdeep.drop_duplicates(subset='Player', keep='first')
bestszn.head(10)
```
