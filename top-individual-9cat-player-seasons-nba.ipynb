```python
ttl = pd.read_csv('/kaggle/input/nba-aba-baa-stats/Player Totals.csv')
print(ttl.head())
print(ttl.columns)
```

```python
cat9 = ttl[['seas_id', 'season', 'player', 'pos', 'tm', 'g', 'pts', 'trb', 'ast', 'stl', 'blk', 'x3p', 'fg_percent', 'ft_percent', 'tov']]
cat9.head()
```

```python
cat9['score'] = cat9['pts'] + cat9['trb'] + cat9['ast'] + cat9['stl'] + cat9['blk'] + cat9['x3p'] + cat9['fg_percent'] + cat9['ft_percent'] - cat9['tov']
sort = cat9.sort_values(by='score', ascending=False)
sort.head(20)
```

```python
best_szn = sort.drop_duplicates(subset='player', keep='first')
best_szn.head(10)
```
