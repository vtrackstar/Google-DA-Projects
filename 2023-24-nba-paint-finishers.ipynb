## Prepare and Process

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
pbp = pd.read_csv('/kaggle/input/nba-2023-24-statss/Untitled spreadsheet - Play by Play.csv')
sht = pd.read_csv('/kaggle/input/nba-2023-24-statss/Untitled spreadsheet - Shooting and Dist.csv')
ttl = pd.read_csv('/kaggle/input/nba-2023-24-statss/Untitled spreadsheet - Totals.csv')

print(pbp.head())
print(sht.head())
print(ttl.head())

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
pbp.rename(columns={'And1\n▼': 'And1'}, inplace=True)
ttl.rename(columns={'FGA\n▼': 'FGA'}, inplace=True)

print(pbp.columns)
print(sht.columns)
print(ttl.columns)
```

```python
df1 = pbp[['Player', 'Pos', 'G', 'MP', 'Off. Fouls Cmt', 'And1', 'Blkd']]
df2 = sht[['Player', 'Dist.', '%FGA 0-3', '%FGA 3-10', 'FG% for 0-3', 'FG% for 3-10', '# Dunks']]
df3 = ttl[['Player', 'FGA', 'PTS', '2PA']]

# Merging the selected columns
merge = df1.merge(df2, on='Player').merge(df3, on='Player')
print(merge.head())
print(merge.columns)
```

```python
merge['Paint Att'] = merge['FGA'] * (merge['%FGA 0-3'] + merge['%FGA 3-10'])
merge['Paint Att No Dnk'] = merge['Paint Att'] - merge['# Dunks']
merge['Paint %'] = merge['FG% for 0-3'] * merge['FG% for 3-10']

merge.head()
```

```python
print(merge.dtypes)
print(merge['Pos'].unique())
print(merge['Pos'].value_counts())
```

## Analyze

```python
# Create a new DataFrame with only Guards
guards = merge[merge['Pos'].isin(['PG', 'SG', 'PG-SG'])]

# Display the new table
guards.head()
```

```python
guards['Paint Score'] = ((guards['And1'] * 1.75) +
                        (guards['Off. Fouls Cmt'] * 0.5) + 
                        (guards['Paint Att No Dnk'] * 1.2) +
                        (guards['Paint %'] * 2) -
                        (guards['# Dunks'] / guards['Paint Att']))


# Sort by Paint Score in descending order
guards_sorted = guards.sort_values(by='Paint Score', ascending=False)

# Display the sorted DataFrame
guards_sorted.head(10)
```

```python
kyrie = guards[guards['Player'] == 'Kyrie Irving']
ant = guards[guards['Player'] == 'Anthony Edwards']

print(kyrie)
print(ant)
```
