```python
df1 = sht[['seas_id', 'player', 'pos', 'tm', 'season', 'mp', 'avg_dist_fga', 'percent_fga_from_x0_3_range', 'percent_fga_from_x3_10_range', 'fg_percent_from_x0_3_range', 'fg_percent_from_x3_10_range', 'num_of_dunks']]
df2 = tot[['seas_id', 'fga']]
df3 = pbp[['seas_id', 'offensive_foul_committed', 'and1', 'shooting_foul_drawn', 'fga_blocked']]

merge = df1.merge(df2, on='seas_id').merge(df3, on='seas_id')
merge.head()
```

```python
merge['paint_fga'] = merge['fga'] * (merge['percent_fga_from_x0_3_range'] + merge['percent_fga_from_x3_10_range'])
merge['paint_fga_no_dunk'] = merge['paint_fga'] - merge['num_of_dunks']
merge['paint_fg%'] = merge['fg_percent_from_x0_3_range'] * merge['fg_percent_from_x3_10_range']

merge.head()
```

```python
merge['paint_score'] = (merge['paint_fga_no_dunk'] * 1.8) + (merge['paint_fg%'] * 3.5) + (merge['and1'] * 2.75) + (merge['offensive_foul_committed'] * -2) + (merge['shooting_foul_drawn'] * 0.5) + (merge['num_of_dunks'] * -2.75)

sort = merge.sort_values(by='paint_score', ascending=False)

sort.head(25)
0
```

```python
best_player_szn = sort.drop_duplicates(subset='player', keep='first')

best_player_szn.head(10)
```

```python
curry = best_player_szn[best_player_szn['player'] == 'Stephen Curry']
ky = best_player_szn[best_player_szn['player'] == 'Kyrie Irving']

print(curry)
print(ky)
```

```python
best_player_szn['shots_in_paint_%'] = best_player_szn['paint_fga'] / best_player_szn['fga']
player_pos = pbp[['seas_id', 'pg_percent', 'sg_percent', 'sf_percent', 'pf_percent', 'c_percent']]

combined = best_player_szn.merge(player_pos, on='seas_id')

comb = combined.fillna(0)

threshold = 3.9

df_above_threshold = comb[comb['c_percent'] > threshold]
df_below_threshold = comb[comb['c_percent'] <= threshold]

col = ['player', 'pos', 'season', 'tm', 'shots_in_paint_%', 'shooting_foul_drawn', 'offensive_foul_committed', 'and1', 'num_of_dunks', 'paint_fga_no_dunk', 'paint_fg%', 'paint_score']

print("Paint Players:")
print(df_above_threshold[col].head(10))

print("Perimeter Players:")
print(df_below_threshold[col].head(10))
```
