```python
sht = pd.read_csv('/kaggle/input/nba-aba-baa-stats/Player Shooting.csv')
ttl = pd.read_csv('/kaggle/input/nba-aba-baa-stats/Player Totals.csv')
print(sht.head())
print(sht.columns)
```

```python
print(ttl.head())
print(ttl.columns)
```

```python
sht_c = sht[['seas_id', 'season', 'player', 'pos', 'age', 'tm', 'g', 'mp', 'fg_percent', 'avg_dist_fga',
       'percent_fga_from_x2p_range', 'percent_fga_from_x0_3_range',
       'percent_fga_from_x3_10_range', 'percent_fga_from_x10_16_range',
       'percent_fga_from_x16_3p_range', 'percent_fga_from_x3p_range',
       'fg_percent_from_x2p_range', 'fg_percent_from_x0_3_range',
       'fg_percent_from_x3_10_range', 'fg_percent_from_x10_16_range',
       'fg_percent_from_x16_3p_range', 'fg_percent_from_x3p_range',
       'percent_assisted_x2p_fg', 'percent_assisted_x3p_fg',
       'percent_dunks_of_fga', 'num_of_dunks', 'percent_corner_3s_of_3pa',
       'corner_3_point_percent']]
ttl_c = ttl[['seas_id', 'birth_year', 'pos', 'age', 'fg', 'fga', 'fg_percent',
       'x3p', 'x3pa', 'x3p_percent', 'x2p', 'x2pa', 'x2p_percent',
       'e_fg_percent', 'ft', 'fta', 'ft_percent', 'pts']]
merge = pd.merge(sht_c, ttl_c, on='seas_id', how='inner')
print(merge.head())
print(merge.columns)
```

```python
calc = merge.copy()
calc['percent_fga_10_3p'] = calc['percent_fga_from_x3p_range'] + calc['percent_fga_from_x16_3p_range'] + calc['percent_fga_from_x10_16_range']
calc['fga_10_3p'] = calc['fga'] * (calc['percent_fga_from_x10_16_range'] + calc['percent_fga_from_x16_3p_range'] + calc['percent_fga_from_x3p_range'])
calc['points_10_3p'] = (((calc['percent_fga_from_x16_3p_range'] * calc['fga']) * calc['fg_percent_from_x16_3p_range'] * 2) + ((calc['percent_fga_from_x10_16_range'] * calc['fga']) * calc['fg_percent_from_x10_16_range'] * 2) + ((calc['percent_fga_from_x3p_range'] * calc['fga']) * calc['fg_percent_from_x3p_range'] * 3))
calc['fg_percent_10_3p'] = (((calc['percent_fga_from_x16_3p_range'] * calc['fga']) * calc['fg_percent_from_x16_3p_range']) + ((calc['percent_fga_from_x10_16_range'] * calc['fga']) * calc['fg_percent_from_x10_16_range']) + ((calc['percent_fga_from_x3p_range'] * calc['fga']) * calc['fg_percent_from_x3p_range'])) / calc['fga']

calc.head()
```

```python
chart = calc[['seas_id', 'season', 'player','percent_fga_from_x2p_range', 'percent_fga_from_x3p_range', 'percent_fga_from_x0_3_range',
       'percent_fga_from_x3_10_range', 'percent_fga_from_x10_16_range',
       'percent_fga_from_x16_3p_range', 'percent_fga_from_x3p_range', 'percent_corner_3s_of_3pa', 'fta', 'ft_percent']]
chart.head()
```

```python
calc_c = calc[['seas_id', 'mp', 'tm', 'g', 'avg_dist_fga', 'percent_fga_10_3p', 'fga_10_3p', 'points_10_3p', 'fg_percent_10_3p']]
grade = pd.merge(calc_c, chart, on='seas_id', how='inner')
grade.head()
```

```python
grade['score'] = (grade['percent_fga_10_3p'] * 1.5) + (grade['points_10_3p'] * 1.25) + (grade['ft_percent'] * 10) + (grade['fg_percent_10_3p'] * 20) + (grade['avg_dist_fga'] * 10)
sort = grade.sort_values(by='score', ascending=False)
col = ['season', 'tm', 'g', 'player', 'mp', 'avg_dist_fga', 'percent_fga_10_3p', 'fga_10_3p', 'points_10_3p', 'fg_percent_10_3p', 'ft_percent', 'score']
sort[col].head(20)
```

```python
best_player_szn = sort.drop_duplicates(subset='player', keep='first')
best_player_szn[col].head(10)
```
